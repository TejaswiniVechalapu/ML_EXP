```python
import pandas as pd
data=pd.read_csv("/content/drive/MyDrive/play_golf_dataset (1).csv")
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
cat_features=['Outlook','Temperature','Humidity','Wind']
data_one=pd.get_dummies(data,columns=cat_features)
```

```python
data_one
```

```python
label=LabelEncoder()
data_one['Play Golf']=label.fit_transform(data_one['Play Golf'])
```

```python
data_one
```

```python
y=data_one['Play Golf']
x=data_one.drop(['Play Golf'],axis=1)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
	x, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
data_g=GaussianNB()
data_g.fit(X_train,Y_train)
```

```python
pre=data_g.predict(X_test)
pre
```

```python
acc=accuracy_score(Y_test,pre)
con=confusion_matrix(Y_test,pre)
cla=classification_report(Y_test,pre)
print(acc)
print(con)
print(cla)
```

```python
pre1=t_g.predict(x_test)
pre1
```

```python
acc1=accuracy_score(y_test,pre1)
con1=confusion_matrix(y_test,pre1)
cla1=classification_report(y_test,pre1)
print(acc1)
print(con1)
print(cla1)
```
